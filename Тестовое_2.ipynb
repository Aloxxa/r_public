1)	Создать запрос, возвращающий топ10 клиентов (customers), отсортированный от большего к меньшему по виду:
a.	customer_id [int]
b.	email [varchar]
c.	payments_count [int]
d.	total_amount [float]




```python
sql_1 = '''

with CustomerPayments as (
    select c.customer_id, c.email,
        count(p.payment_id) as payments_count,
        sum(p.amount) as total_amount,
        dense_rank() over (order by count(p.payment_id) desc, sum(p.amount) desc) as customer_rank
    from customers c
    left join payment p ON c.customer_id = p.customer_id
    group by c.customer_id, c.email
)
select cp.customer_id, cp.email, cp.payments_count, cp.total_amount
from CustomerPayments cp
where customer_rank <= 10
order by customer_rank, cp.customer_id desc, cp.email desc

'''
#dense_rank если будет одинаковая sum(p.amount), то выведутся все с этой sum(p.amount)
```

```python
select(sql_1)
```

2)	Создать запрос, возвращающий количество фильмов, максимальную и минимальную длину (length) фильма, средний рейтинг (rating) и он же разбитый на группы (9–10, 7-8, 5-6 и менее 5) для всех фильмов, в которых снимался актер с фамилией “James

```python
sql_2 = '''
with CTE_1 as (
    select
        f.film_id,
        f.length,
        f.rating
    from film f
    left join film_actor fa on f.film_id = fa.film_id
    left join actor a on fa.actor_id = a.actor_id
    where a.last_name = 'James'
)
select
    count(*) as total_films,
    max(length) as max_length,
    min(length) as min_length,
    avg(rating) as avg_rating,
    case
        when avg(rating) between 9 and 10 then '9-10'
        when avg(rating) between 7 and 8 then '7-8'
        when avg(rating) between 5 and 6 then '5-6'
        else '<5'
    end as rating_group
from CTE_1
'''
```

3)	Найти по 3 города (city) c наибольшей и наименьшей выручкой от платежей клиентов (amount) за последние 30 дней (payment_date)

```python
sql_3_max = '''

select city
from (
    select c.city,
        sum(p.amount) as total_amount,
        dense_rank() over (order by sum(p.amount) desc) as dr
    from city c
    left join address a on c.city_id = a.city_id
    left join staff s on a.address_id = s.address_id
    left join payment p on s.staff_id = p.staff_id
    where p.payment_date >= date('now', '-30 days')
    group by c.city
) as t1
where dr <= 3

'''
```

```python
select(sql_3_max)
```

```python
sql_3_min = '''

select city
from (
    select c.city,
        sum(p.amount) as total_revenue,
        dense_rank() over (order by sum(p.amount) asc) as dr
    from city c
    left join address a on c.city_id = a.city_id
    left join staff s on a.address_id = s.address_id
    left join payment p on s.staff_id = p.staff_id
    where p.payment_date >= date('now', '-30 days')
    group by c.city
) as t1
where dr <= 3

'''
```

```python
select(sql_3_min)
```

5)	Показать топ 3 категории фильмов (name) c топ5 фильмами внутри категории по изменению рейтинга фильма (рейтинг) за последний месяц ( данные есть за каждый день в поле last_update)

```python
sql_5 = '''

with TopCategories AS (
    select c.name AS category_name, f.title, f.rating,
        dense_rank() over (partition by c.name order by f.rating desc) as rank
    from film f
    left join film_category fc ON f.film_id = fc.film_id
    left join category c ON fc.category_id = c.category_id
    left join (
        select c.name AS category_name,
            sum(f.rating) as total_rating,
            dense_rank() OVER (order by AVG(f.rating) DESC) AS rank
        FROM film f
        left join film_category fc ON f.film_id = fc.film_id
        left join category c ON fc.category_id = c.category_id
        group by c.name
    ) AS top_categories on c.name = top_categories.category_name
    where top_categories.rank <= 3
        and f.last_update >= DATE('now', '-1 month')
)
select category_name, title, rating
from TopCategories
where rank <= 5
order by category_name, rating DESC;

'''
#Если убрать and f.last_update >= DATE('now', '-1 month'), то есть значения
#Верятно, елси изменить интервал в film_df (сейчас 5 лет), то все ок будет
#Запрос выводит пустую таблицу - 50% сделано - стакан на половину полон
```
